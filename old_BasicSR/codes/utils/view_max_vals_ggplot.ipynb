```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from cycler import cycler
# import ggplot
# from plotnine import *
```

```python
pth_in='/data_dir/scene-lists/order-1-stats-max.txt'
pth_in='/mnt/g/Planet/api/hist_stats/stats_39.xlsx'
df=pd.read_excel(pth_in, sheet_name=1)
df.head()
```

```python
# use grouping
f, ax = plt.subplots(3)
dfb=df.groupby('Band')
stat=['MIN','MEAN','MAX']
# dfb.rem
for i in [0,1,2]:
    dfb.plot.hist(y=stat[i],ax=ax[i], bins=20, alpha=0.5)
    # plt.rc('lines', linewidth=4)
    plt.rc('axes', prop_cycle=(cycler('color', ['y', 'r', 'g', 'b']))) #+cycler('linestyle', ['-', '--', ':', '-.'])))
    ax[i].legend().remove()
    ax[i].set_title(stat[i])
ax[0].legend(['B','G','R','N'])
```
