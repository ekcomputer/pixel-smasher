## Viewing planet subsets

To check bands, ratios, histograms

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from cycler import cycler
import os
from IPython.display import display, Image
from skimage import data
from skimage.filters import try_all_threshold # better than cv2
```

```python
img_list=[]
pth_in='/mnt/g/Planet/subsets_practice/'
for root, _, file_list in sorted(os.walk(pth_in)): # +'/*SR.tif'
        for x in file_list:  # assume only images in the input_folder
            if x.endswith(".png"):
                path = os.path.join(root, x) 
                img_list.append(path)
```

```python
f, ax = plt.subplots(1,2)
b1=0
b2=1
for file in img_list[2:3]:
    img=plt.imread(file)
    ndwi=(img[:,:,b2]-img[:,:,b1])/(img[:,:,b2] + img[:,:,b1])
    plt.colorbar
#     plt.imshow(img) +     plt.hist(img[:,:,2])
    ax[0].imshow(ndwi, cmap='Greys')
    ax[1].imshow(img)
#     Image(file)
```

```python
fig, ax = try_all_threshold(ndwi, figsize=(10, 8), verbose=False)
plt.show()
```
